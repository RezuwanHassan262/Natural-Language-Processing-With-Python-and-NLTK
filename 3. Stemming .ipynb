Stem: Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a **lemma**. 

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
```

```python
ps = PorterStemmer()
```

```python
example_words = ["python","pythoneer","pythoning","pythoned","pythonly"]
for w in example_words:
  print(ps.stem(w))
```

```python
new_text = "It is very important to be pythonly while you are pythoning with python. All pythoneers have pythoned poorly at least once."

words = word_tokenize(new_text)

for w in words:
	print(ps.stem(w))
```
